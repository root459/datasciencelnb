```python
df=data[[
    'customerID', 'gender', 'SeniorCitizen', 'Partner', 'Dependents',
    'tenure', 'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
    'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
    'PaymentMethod', 'MonthlyCharges', 'TotalCharges', 'Churn'
]].replace({'Yes':1,'No':0,'No phone service':3,'Female':0,'Male':1,'DSL':0,'Fiber optic':1,'Electronic check':3, 'Mailed check':4, 'Bank transfer (automatic)':5,
       'Credit card (automatic)':6,'Month-to-month':7, 'One year':8, 'Two year':9,'No internet service':10})
```

```python
df.drop('customerID',axis=1,inplace=True)
```

```python
#df['TotalCharges'] = df['TotalCharges'].replace(' ',0)
#df['TotalCharges'] = df['TotalCharges'].replace('',0)
#df['TotalCharges'] = df['TotalCharges'].astype(float)
#df[df['TotalCharges']==' ']
df = df[df['TotalCharges'] != ' ']
df['TotalCharges']=df['TotalCharges'].astype(float)
df.info()
```

```python
X=df[['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'tenure', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod','MonthlyCharges']]
Y=df['TotalCharges']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.05, random_state=42)
reg = LinearRegression()
reg.fit(X_train,Y_train)
predict=reg.predict(X_test).round(2)
predict
```
